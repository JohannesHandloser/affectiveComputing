```python
from sklearn.preprocessing import StandardScaler

data_only_max_scaled = MinMaxScaler().fit_transform(data_frame.iloc[:, :16])
data_only_standard_scaled = StandardScaler().fit_transform(data_frame.iloc[:, :16])
label_only = data_frame.iloc[:, 16:]
```

```python
import pandas as pd 

pd.plotting.scatter_matrix(data_frame, alpha = 0.1, figsize = (20,20), diagonal = 'kde');
```

```python
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca = pca.fit(data_only_max_scaled)

reduced_data_max = pca.transform(data_only_max_scaled)
reduced_data_max = pd.DataFrame(reduced_data_max, columns = ['Dimension 1', 'Dimension 2'])

plt.scatter(reduced_data_max['Dimension 1'], reduced_data_max['Dimension 2'], c=label_only, label=label_only)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.rcParams["figure.figsize"] = (20,20)
plt.show();
```

```python
pca_standard = PCA(n_components=2)
pca_standard = pca.fit(data_only_standard_scaled)

reduced_data_standard = pca_standard.transform(data_only_standard_scaled)
reduced_data_standard = pd.DataFrame(reduced_data_standard, columns = ['Dimension 1', 'Dimension 2'])

plt.scatter(reduced_data_standard['Dimension 1'], reduced_data_standard['Dimension 2'], c=label_only, label=label_only)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.rcParams["figure.figsize"] = (20,20)
plt.show();
```
